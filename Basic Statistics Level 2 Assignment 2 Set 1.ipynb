```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
measure_X=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
name_of_company=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
# Pie Plot
plt.figure(figsize=(8,8))
plt.pie(measure_X,labels=name_of_company,autopct='%1.2f%%')
plt.title("Pie Chart")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(measure_X)
plt.title("Box Plot")
plt.show()
```

```python
# To find Mean
print('Mean = ',measure_X.mean())

# To find Median
print('Median =',measure_X.median())

# To find Variance
print('Variance = ',measure_X.var())

# To find Standard Deviation
print('Standard Deviation= ',measure_X.std())
```
